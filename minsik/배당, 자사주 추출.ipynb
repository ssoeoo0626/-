```python
import warnings

# 모든 경고를 무시
warnings.filterwarnings('ignore')

회사명 = '진양홀딩스'  # 실제 회사명으로 대체해야 합니다.
start_year = 2015
end_year = 2022  # 15년부터 23년까지 데이터를 얻고 싶다면 2023년까지 설정

# 현금흐름표 데이터를 저장할 빈 DataFrame 생성
all_cash_flows = pd.DataFrame()

# 지정한 연도 범위에 대한 데이터 수집
for year in range(start_year, end_year + 1):
    cf = dart.finstate_all(회사명, year, reprt_code='11011', fs_div='CFS')
    
    # 현금흐름표 (sj_div이 'CF'인 데이터) 필터링
    cf2 = cf[cf['sj_div'].str.contains("CF")]
    
    # '자기주식' 또는 '배당금'이 포함된 행 추출 (두 조건 중 하나라도 충족하는 행)
    자기주식_또는_배당금_df = cf2[cf2['account_nm'].str.contains("자기주식|배당금")]
    
    # 'thstrm_amount' 열의 값에서 숫자와 쉼표만 남기고 나머지 문자 제거
    자기주식_또는_배당금_df['thstrm_amount'] = 자기주식_또는_배당금_df['thstrm_amount'].str.replace('[^0-9,]', '', regex=True)
    
    # 'thstrm_amount' 열의 값 숫자로 변환
    자기주식_또는_배당금_df['thstrm_amount'] = pd.to_numeric(자기주식_또는_배당금_df['thstrm_amount'])
    
    # 'thstrm_amount' 열에 천 단위 구분자 추가
    자기주식_또는_배당금_df['thstrm_amount'] = 자기주식_또는_배당금_df['thstrm_amount'].apply(lambda x: f'{x:,}')
    
    # 필요한 열만 선택하여 데이터프레임 연결
    selected_columns = 자기주식_또는_배당금_df[['bsns_year', 'sj_nm', 'account_nm', 'thstrm_nm', 'thstrm_amount']]
    
    # 데이터를 all_cash_flows에 추가
    all_cash_flows = pd.concat([all_cash_flows, selected_columns], ignore_index=True)
    
all_cash_flows
```
