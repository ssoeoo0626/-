```python
import time
import warnings
from tqdm import tqdm
# 모든 경고를 무시
warnings.filterwarnings('ignore')

# 기업 코드 목록을 가져오기



corp_names = [company['corp_name'] for company in filtered_data]
#0~3636까지 있음. 
#https://yenpa.tistory.com/49 
# 특정 순번의 회사 뽑기
start_idx =20  #시작
end_idx =100    #종료          
selected_companies = corp_names[start_idx:end_idx]

# 데이터를 저장할 빈 DataFrame 생성
total_share = pd.DataFrame()

start_year=2015
end_year=2022

# 선택된 회사들에 대해 데이터 추출
for 회사명 in tqdm(selected_companies, desc="진행 중"):
    for year in range(start_year, end_year + 1):
        try:
            # 데이터를 가져오는 코드
            share = dart.report(회사명, '주식총수', year)
            
            # 필요한 열만 선택하여 데이터프레임 생성
            selected_columns = share[['corp_name', 'se', 'isu_stock_totqy', 'now_to_isu_stock_totqy', 'now_to_dcrs_stock_totqy',
                                      'redc', 'profit_incnr', 'istc_totqy', 'tesstk_co', 'distb_stock_co']]
            
            # 데이터프레임의 열 이름 변경
            selected_columns.columns = ['회사명', '구분', '발행주식 총수', '현재까지 발행한 주식의 총수', '현재까지 감소한 주식의 총수',
                                        '감자', '이익소각', '발행주식의 총수', '자기주식수', '유통주식수']
            
            # '연도' 열 추가
            selected_columns['연도'] = year
            
            # 데이터를 total_share에 추가
            total_share = pd.concat([total_share, selected_columns], ignore_index=True)
            
            # '구분' 열에서 '합계' 문자열을 포함한 행 추출
            filtered_total_share = total_share[total_share['구분'].str.contains('합계')]
            
            # 딜레이 추가 (예: 1초 딜레이)
            time.sleep(0.5)  # 1초의 딜레이를 추가합니다
            

        except KeyError:
            # 필요한 열이 없는 경우 해당 회사를 건너뛰기
            #print(f"Skipping {회사명} for year {year}")
            continue

# 엑셀 파일로 저장 (동적 파일명 사용)
excel_file_name = f'filtered_share_data_{start_idx}_{end_idx}.xlsx'
filtered_total_share.to_excel(excel_file_name, index=False)
print(f"Data saved to {excel_file_name}")
```

```python
import pandas as pd
import glob

folder_path = 'D:/stock/filtered_share_data'
excel_files = glob.glob(folder_path + '/*.xlsx')  # 확장자가 .xlsx인 파일을 모두 가져옴

dfs = []  # 데이터프레임을 저장할 리스트 생성

for file in excel_files:
    df = pd.read_excel(file)  # 엑셀 파일을 데이터프레임으로 읽음
    dfs.append(df)  # 데이터프레임을 리스트에 추가

# 모든 데이터프레임을 하나의 데이터프레임으로 합침
combined_df = pd.concat(dfs, ignore_index=True)


```
