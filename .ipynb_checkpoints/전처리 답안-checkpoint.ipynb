### 공기의 질을 측정한 데이터
1. airquanlity 데이터에 대해서 결측치가 가장 많은 변수를 찾아서 해당 결측치를 0으로 대체하고 결측치를 제외한 평균과 0으로 대치한 후 평균의 차이를 구하시오
2. wind 변수에 대해서 Min-Max 정규화를 수행 한 후 평균 값과 Z 정규화를 수행 한 후 평균 값의 차이를 구하시오
3. 월별 평균 기온을 구하시오

```python
import pandas as pd
import numpy as np

# csv 파일이 위치한 디렉토리 입력
df = pd.read_csv('data/airquality.csv')
df.head()
```

```python
## info() 함수를 이용하여 결측치의 갯수를 파악
df.info()
```

```python
# Ozone 컬럼의 결측치를 0으로 대체
# 대체 이전과 이후의 평균을 출력
print(df["Ozone"].isnull().sum())         # Ozone 컬럼의 결측치 개수 출력
Ozone_mean = df["Ozone"].mean()           # Ozone 컬럼의 평균

df["Ozone"].fillna(0, inplace=True)       # Ozone 컬럼의 결측치를 평균으로 대체

print(df["Ozone"].isnull().sum())         # 대체 후 Ozone 컬럼의 결측치 개수 출력
Ozone_mean2 = df["Ozone"].mean()          # 체치 후 Ozone 컬럼의 평균

print(Ozone_mean)
print(Ozone_mean2)
```

```python
## 두 평균 값의 차이 출력
print(Ozone_mean-Ozone_mean2)
```

```python
# Min-Max 표준화, Z 표준화 변환뒤 컬럼 추가
# 소수점 2자리까지 표시
Min = np.min(df["Wind"])
Max = np.max(df["Wind"])
df["min_max"] = round((df["Wind"] - Min)/(Max - Min), 2)

Mean = np.mean(df["Wind"])
Std = np.std(df["Wind"])
df["z"] = round((df["Wind"] - Mean)/Std, 2)

## Min-Max 표준화는 연속형 변수의 값을 0과 1사이의 값으로 변환. 스케일이 다른 두 변수를 Min-Max 변환하면 상호간에 비교가 가능
## z 표준화는 변수값에서 평균을 뺀 후 표준편차로 나누는 것으로, 변환 후 데이터의 평균은 0, 표준편차는 1
```

```python
df.describe()
```

- Min-Max 표준화된 Wind 컬럼의 평균은 0.433, Z 표준화 된 컬럼의 평균은 0.000으로 그 차이는 0.433

```python
# 월별 평균기온
print(df.groupby("Month")['Temp'].mean())
```

### 근무자 정보를 측정한 데이터
1. Attrition은 퇴사 여부 데이터를 (yes:1, no:0)으로 데이터를 변환하여 새로운 파생변수에 대입하고 범주별 레코드 수를 계산하시오
2. 범주형 변수 중 유일한 값을 1개만 가지고 있는 컬럼을 찾아내어 그 컬럼을 데이터에서 제거하시오
3. 원래 데이터에서 숫자형인 컬럼만 추출하여 새로운 데이터프레임으로 생성하고, 각 변수간의 상관계수(피어슨 상관계수)를 구하시오

```python
# 필요한 파이썬 라이브러리 불러오기
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# csv 파일이 위치한 디렉토리 입력
df = pd.read_csv('data/HR-Employee-Attrition.csv')
df.head()
```

```python
## 퇴사 여부 데이터를 (yes:1, no:0)으로 데이터를 변환하여 새로운 파생변수에 대입하고 범주별 레코드 수를 계산
target_map = {'Yes':1, 'No':0}

df["Attrition_numerical"] = df["Attrition"].apply(lambda x: target_map[x])
df["Attrition_numerical"].value_counts()
```

```python
df.info()
```

```python
## 범주형 변수 중 유일한 값을 1개만 가지고 있는 컬럼을 찾아내어 그 컬럼을 데이터에서 제거
cat_feat = df.select_dtypes('object','category').columns.values
df_cat = df[cat_feat].copy()
df_cat.nunique().sort_values()
```

```python
## 범주형 변수 중 유일한 값을 1개만 가지고 있는 컬럼은 Over18 이다
df_cat = df_cat.drop(['Over18'], axis=1, errors='ignore')
```

```python
## 숫자형인 컬럼만 추출하여 새로운 데이터프레임으로 생성
num_feat = df.select_dtypes('number').columns
attrition_num = df[num_feat].copy().corr(method='pearson')
```
